### Let's Get Started

Import Necessary Library

```python
from seleniumwire import webdriver
from seleniumwire.utils import decode
import json
import pandas as pd
```

set a new instance of the chrome driver

```python
driver = webdriver.Chrome()
```

set url

```python
url = 'https://shopee.co.id/search?keyword=mechanical%20keyboard'
driver.get(url)
```

Create requests variable

```python
for request in driver.requests:
    if request.response:
        if request.url.startswith('https://shopee.co.id/api/v4/search/search_items?by=relevancy&keyword=mechanical%20keyboard&limit=60&newest=0&order=desc&page_type=search&scenario=PAGE_GLOBAL_SEARCH&version=2'):
            #step1, if success try step2
            # body=request.body
            # print(body)

            # step2, if success try step3
            # response=request.response
            # body=decode(response.body, response.headers.get('Content-Encoding', 'Identify'))
            # print (body)

            # Step3, if success go to last step
            # response=request.response
            # body=decode(response.body, response.headers.get('Content-Encoding', 'Identify'))
            # decoded_body = body_decode('utf')
            # print(decoded_body)

            #last step
            response = request.response
            body = decode(response.body, response.headers.get('Content-Encoding', 'Identify'))
            decoded_body = body.decode('utf8')
            json_data = json.loads(decoded_body)
            
            #test 1 product
            product_name = json_data['items'][0]['item_basic']['name']
```

```python
print (product_name)
```

Check total product rows in 1 page

```python
rows = json_data['items']
print (len(rows))
```

Scrape the data & input to array

```python
data=[]
for i in range (0, len(rows)):
    #general info
    item_id       = json_data['items'][i]['item_basic']['itemid']
    name          = json_data['items'][i]['item_basic']['name']
    harga         = int(json_data['items'][i]['item_basic']['price']/100000)
    stock         = int(json_data['items'][i]['item_basic']['stock'])
    price_max     = int(json_data['items'][i]['item_basic']['price_max']/100000)
    price_min     = int(json_data['items'][i]['item_basic']['price_min']/100000)
    sold          = int(json_data['items'][i]['item_basic']['sold'])
    shop_location = json_data['items'][i]['item_basic']['shop_location']
    shop_id       = json_data['items'][i]['item_basic']['shopid']

    #ratings
    stars    = json_data['items'][i]['item_basic']['item_rating']['rating_star']
    star     ='%.2f' %stars #convert integer into 2 digit
    rating_1 = json_data['items'][i]['item_basic']['item_rating']['rating_count'][1]
    rating_2 = json_data['items'][i]['item_basic']['item_rating']['rating_count'][2]
    rating_3 = json_data['items'][i]['item_basic']['item_rating']['rating_count'][3]
    rating_4 = json_data['items'][i]['item_basic']['item_rating']['rating_count'][4]
    rating_5 = json_data['items'][i]['item_basic']['item_rating']['rating_count'][5]

    #test the output:
    #print(item_id, name, price, stock, price_max, price_min, sold, shop_location, shop_id, star, rating_1, rating_2, rating_3, rating_4, rating_5)

    # if success, next step : input to array data[]
    data.append(
        (item_id, name, harga, stock, price_max, price_min, sold, shop_location, shop_id, star, rating_1, rating_2, rating_3, rating_4, rating_5)
    )

#input to pandas dataframe
df = pd.DataFrame(data, columns=['item_id', 'Name', 'Price', 'Stock', 'price_max', 'price_min', 'Sold', 'shop_location', 'shop_id', 'stars', 'rating_1', 'rating_2', 'rating_3', 'rating_4', 'rating_5'])
df #check the dataframe
```
